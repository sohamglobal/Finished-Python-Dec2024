**Creating numpy from random numbers**

```python
import numpy
```

```python
arr=numpy.random.random(10)
print(arr)
```

```python
arr=numpy.random.random((3,5))
print(arr)
```

```python
arr=numpy.random.randint(15)
print(arr)
```

```python
arr=numpy.random.randint(1,45,10)
print(arr)
```

```python
x=numpy.random.choice(arr)
print(x)
```

```python
arr=numpy.random.randint(1,45,(3,4))
print(arr)
```

```python
print(arr.dtype)
```

**numpy array of other data types**

```python
arrs=numpy.array(['praffull','java','python','sql','spark','microsoft','volkswagen'])
print(arrs)
```

```python
print(arrs.dtype)
```

```python
arrd=numpy.arange('2024-12-01','2025-01-05',dtype='datetime64[D]')
print(arrd)
```

```python
print(arrd.dtype)
```

```python
arrb=numpy.array([True,True,False,True,False,False,False,True])
print(arrb)
```

```python
print(arrb.dtype)
```

```python
import numpy
```

```python
arr=numpy.array([13,9,26,1,30,23,45])
```

```python
print(numpy.average(arr))
```

```python
print(numpy.mean(arr))
print(numpy.median(arr))
print(numpy.corrcoef(arr))
print(numpy.var(arr))
print(numpy.amax(arr))
print(numpy.amin(arr))
```

```python
print(numpy.sin(arr))
```

**Creating numpy array using single dimensional list**

```python
lst=[9,26,13,1,30,45,5,11,75,25]
print(type(lst))
print(lst)
arr=numpy.array(lst)
print(type(arr))
print(arr)
```

**Creating numpy array from a double dimensional list**

```python
lst=[
    [12,23,34,66],
    [65,43,21,33],
    [76,89,58,22]
]
print(lst)
arr=numpy.array(lst)
print(arr)
```

```python
res=arr+3
print(res)
```

```python
sqr=arr**2
print(sqr)
```

**Creating numpy array from range of values**

```python
arr=numpy.arange(45,60)
print(arr)
arr=numpy.arange(1,25,3)
print(arr)
arr=numpy.arange(75,50,-4)
print(arr)
```

**Creating numpy array from Linear spacing**

```python
arr=numpy.linspace(1,13,9)
print(arr)
```

**Numpy arrays with zeros and ones**

```python
arr0=numpy.zeros((4,6))
print(arr0)
arr1=numpy.ones((5,8))
print(arr1)
```

**Subset from 1-D NumPy array**

```python
import numpy
```

```python
arr=numpy.array([12,23,34,45,56,67,78,89,90])
```

```python
print(arr)
```

```python
print(arr[5])
```

```python
print(arr[2:7])
```

```python
print(arr[3:])
print(arr[:5])
print(arr[-1])
```

```python
print(arr[arr%2==1])
print(arr[arr>50])
```

**Subset from 2-D numpy arrays**

```python
arr=numpy.array([
    [34,78,22,12,54],
    [60,16,41,38,99],
    [21,55,82,51,94]
])
print(arr)
```

```python
#all rows all columns
print(arr)
print(arr[:,:])
```

```python
print(arr[:,0:3])
```

```python
print(arr[1,2])
```

```python
print(arr[1:,2:])
```

```python
print(arr[2,:])
```

```python
print(arr[arr%2==1])
print(arr[arr<50])
```

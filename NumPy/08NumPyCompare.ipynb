```python
import numpy
```

```python
a1=numpy.array([[1,2,3],[4,5,6]])
a2=numpy.array([[9,2,5],[3,5,6]])
print(a1)
print(a2)
```

```python
comp=(a1==a2)
print(comp)
```

```python
print(comp.all())
```

```python
if comp.all():
  print('both numpy arrays are exactly same')
else:
  print('numpy arrays are different')
```

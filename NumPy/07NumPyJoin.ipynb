```python
import numpy
```

```python
a1=numpy.array([34,78,52])
a2=numpy.array([12,34,56,78,90,31])
```

```python
res=numpy.concatenate((a1,a2))
print(res)
```

```python
a3=numpy.array([[1,2,3],[4,5,6]])
a4=numpy.array([[8,7,1],[6,2,9],[1,5,3]])
print(a3)
print(a4)
```

```python
a5=numpy.array([[1,2,3],[4,5,6]])
a6=numpy.array([[7,9],[4,6]])
print(a5)
print(a6)
```

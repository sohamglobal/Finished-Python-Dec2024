**retrieving data from Joins**

```python
curs.execute("select customers.custnm,cars.modelnm from customers inner join cars on customers.carid=cars.carid")
```

```python
data=curs.fetchall()
print(data)
```

```python
curs.execute("select customers.custnm,cars.modelnm from customers left outer join cars on customers.carid=cars.carid")
```

```python
data=curs.fetchall()
print(data)
```

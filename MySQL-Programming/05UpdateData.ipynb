**Update employee salary**

**Change employee location**

```python
no=int(input('Enter employee number : '))
newloc=input('Enter new location : ')
cnt=curs.execute("update employees set location='%s' where empno=%d" %(newloc,no))
#print(cnt)
con.commit()
if cnt==1:
  print('location updated successfully')
else:
  print('employee not found')
```

**Deposit amount in bank account**

```python
no=int(input('Enter account number : '))
amt=float(input('Enter deposit amount :'))

cnt=curs.execute("update accounts set balance=balance+%.2f where accno=%d" %(amt,no))
con.commit()
if cnt==1:
  print('balance updated successfully')
else:
  print('account does not exist')
```

**Withdraw amount from bank account**

```python
no=int(input('Enter account number : '))
amt=float(input('Enter withdraw amount :'))

try:
  cnt=curs.execute("update accounts set balance=balance-%.2f where accno=%d" %(amt,no))
  con.commit()
  if cnt==1:
    print('balance updated successfully')
  else:
    print('account does not exist')
except:
  print('insufficient balance')
```

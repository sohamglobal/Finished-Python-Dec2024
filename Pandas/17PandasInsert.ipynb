**Inserting rows in a dataframe**

```python
import pandas
```

```python
df=pandas.DataFrame(columns=['name','science','maths','english'])
print(df)
```

```python
df.loc[0]=['praffull',87,42,76]
print(df)
```

```python
df.loc[1]=['swami',97,93,95]
print(df)
```

```python
df.loc[2,['name','english']]=['ajinkya',75]
print(df)
```

```python
df.loc[3,'maths']=57
print(df)
```

**Insert columns in a dataframe**

```python
dic={
    'country':['england','germany','japan','usa','france'],
    'capital':['london','berlin','tokyo','washington','paris']
}
```

```python
df=pandas.DataFrame(dic)
print(df)
```

```python
df.insert(0,'code',[x for x in range(151,156)])
print(df)
```

```python
lst=['pound','euro','yen','dollar','euro']
df.insert(3,'currency',lst)
print(df)
```

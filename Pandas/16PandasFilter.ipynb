```python
import pandas
```

```python
url="http://sohamspring.azurewebsites.net/api/accounts/all"
df=pandas.read_json(url)
print(df)
```

```python
print(df[df['acctype']=='ppf'])
```

```python
print(df[df['acctype']!='saving'])
```

```python
print(df[df['balance']<45000])
```

```python
print(df[(df['acctype']=='saving') & (df['balance']>80000)])
```

```python
print(df[(df['acctype']=='ppf') | (df['balance']>95000)])
```

```python
print(df[df['accno'].isin([1013,1029,1041,1035])])
```

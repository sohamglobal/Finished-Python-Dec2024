**Data joining in Pandas**

```python
import pandas
```

```python
url1="http://sohamspring.azurewebsites.net/api/customers/all"
dfcust=pandas.read_json(url1)
print(dfcust)
```

```python
url2="http://sohamspring.azurewebsites.net/api/products/all"
dfcars=pandas.read_json(url2)
print(dfcars)
```

**INNER JOIN**

```python
res=dfcust.merge(dfcars,on="prodid",how="inner")
print(res)
```

```python
print(res[['custnm','prodnm','company']])
```

**Left outer join**

```python
res=dfcust.merge(dfcars,on="prodid",how="left")
print(res[['custnm','prodnm','company','price']])
```

**Right outer join**

```python
res=dfcust.merge(dfcars,on='prodid',how='right')
print(res[['custnm','prodnm','company','price']])
```

```python
import pandas
```

```python
url1="http://sohamspring.azurewebsites.net/api/customers/all"
dfcust=pandas.read_json(url1)
print(dfcust)
```

```python
url2="http://sohamspring.azurewebsites.net/api/products/all"
dfcars=pandas.read_json(url2)
print(dfcars)
```

```python
res=dfcust.merge(dfcars,on="prodid",how="left")
res=res[['custnm','prodnm','company','price']]
print(res)
```

```python
print(pandas.isnull(res))
```

**Remove rows with null values**

```python
rowswithoutnull=res.dropna()
print(rowswithoutnull)
```

```python
colswithoutnull=res.dropna(axis=1)
print(colswithoutnull)
```

Replace null with other value

```python
res['prodnm'].fillna('discontinued',inplace=True)
print(res)
```

```python
res=res.replace('Volkswagen','vw')
print(res)
```

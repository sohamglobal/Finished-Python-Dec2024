**Series - Single dim data**

```python
import numpy
import pandas
```

```python
lst=['mumbai','london','berlin','dubai','tokyo','moscow','new york']
```

```python
arr=numpy.array(lst)
print(arr)
```

```python
ser=pandas.Series(arr)
print(ser)
```

```python
print(ser[3])
```

**DataFrame - double dimensional data**

```python
arr2=numpy.array([
    ['the matrix','pk','sholay'],
    [1999,2014,1975],
    ['english','hindi','hindi'],
    ['action','comedy','action'],
    [8.7,8.4,7.9],
    ['keanu reeves','amir khan','amitabh bachchan']
])
print(arr2)
```

```python
print(arr2.shape)
```

```python
df=pandas.DataFrame(arr2)
print(df)
```

```python
rows=['filmname','year','language','genre','imdb','actor']
cols=['film1','film2','film3']
df2=pandas.DataFrame(arr2,rows,cols)
print(df2)
```

```python
print(df2['film2'])
```

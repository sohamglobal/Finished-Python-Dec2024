```python
cho=None
dic={}

while cho!='no':
  co=input('Enter company : ')
  pr=input('Enter product : ')
  if co not in dic:
    dic[co]=[pr]
  else:
    dic[co].append(pr)

  cho=input('any more company? ')
else:
  print(dic)
```

```python
documents={
    "doc1":"cricket is a popular sport",
    "doc2":"formula-1 racing is also a popular sport",
    "doc3":"cricket and formula-1 are both popular"
}

index={}

for docid,content in documents.items():
  words=content.split()
  for word in words:
    if word in index:
      index[word].add(docid)
    else:
      index[word]={docid}


print(index)
```

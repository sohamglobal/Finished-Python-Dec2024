```python
w=None
while w!='stop':
  w=input('Enter a word : ')
  if w in words:
    words[w]+=1
  else:
    words[w]=1

print(words)
```

```python
sen="cricket is popular in india and football is popular worldwide"
wc={}

words=sen.split()
print(words)

for w in words:
  wc[w]=wc.get(w,0)+1

print(wc)
```

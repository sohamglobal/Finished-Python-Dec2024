```python
cho="yes"
countries=[]

while cho.lower()== 'yes':
  nm=input('Enter name : ')
  countries.append(nm)
  cho=input('Continue? (yes/no) ')

print(countries)
```

```python
con=input('Enter a country ')

if con in countries:
  print(con+" found in the list")
else:
  print('not found...')
  cho=input('want to add to the list? (yes/no) ')
  if cho.lower()=='yes':
    countries.append(con)
    print(countries)
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy
```

```python
inputs=numpy.array([-1.0,0.0,1.0,3.0,2.0,5.0,4.0],dtype=float)
outputs=numpy.array([-2.0,1.0,4.0,10.0,7.0,16.0,13.0],dtype=float)
```

```python
model=tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(inputs,outputs,epochs=100)
prediction=model.predict(numpy.array([[9.0],[15.0]]))
print(prediction)
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/Flipkart_Mobiles.csv"
df=pandas.read_csv(url)
print(df)
```

```python
sns.set(style="whitegrid")
```

```python
top_brands=df['Brand'].value_counts().head(5)
print(top_brands)
```

```python
sns.barplot(x=top_brands.values,y=top_brands.index,palette="viridis")
pl.title('Top 5 selling mobile brands')
pl.xlabel('Number of mobiles sold')
pl.ylabel('Brand')
pl.show()
```

```python
top_colors=df['Color'].value_counts().head(5)
sns.barplot(x=top_colors.values,y=top_colors.index,palette="muted")
pl.title('Top 5 selling colors')
pl.xlabel('Number of mobiles sold')
pl.ylabel('Color')
pl.show()
```

**Basic Data visualization using MatPlotLib**

```python
from matplotlib import pyplot
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
sale=[73,42,84,60,49,103]
```

**Line Chart**

```python
pyplot.plot(months,sale,color='red',linestyle='dashdot')
pyplot.xlabel('Months')
pyplot.ylabel('Sale')
pyplot.title('Mobile sale in 2024')
pyplot.show()
```

**Bar chart**

```python
clrs=['green','blue','red','teal','purple','pink']
pyplot.bar(months,sale,color=clrs)
pyplot.show()
```

**Scatter Chart**

```python
pyplot.scatter(months,sale,color='green')
pyplot.show()
```

**Pie chart**

```python
comp=['samsung','apple','moto','realme','oppo']
total=[2419,1724,621,6275,4318]
```

```python
pyplot.pie(total,labels=comp,autopct='%.1f%%')
pyplot.show()
```

**Combine charts**

```python
pyplot.scatter(months,sale)
pyplot.plot(months,sale)
pyplot.bar(months,sale)
pyplot.show()
```
